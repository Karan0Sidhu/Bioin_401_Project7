```python
import requests
from tqdm import tqdm
import xml.etree.ElementTree as ET
import time
import json

# Function to check if a text contains more than one word
def contains_multiple_words(text):
    return len(text.split()) > 100

# URL to fetch data from PubMed
output_file_path = "/content/drive/MyDrive/my_files/PUBMED_ID_CHEMICALNAMES_TEXTS.json"
failed_pmid_file_path = "/content/drive/MyDrive/my_files/Failed_PMIDs.txt"

# Dictionary to store text associated with each PMID
pmid_texts = {}

with open(failed_pmid_file_path, 'w', encoding='utf-8') as failed_file:
    # Initialize tqdm to track progress
    with open('PUBMED_IDs_THAT_NAV_LOVES.txt', 'r', encoding='latin-1') as infile:
        lines = infile.readlines()
        total_lines = len(lines)
        progress_bar = tqdm(total=total_lines)

        start_time = time.time()
        for pmid in lines:
            pmid = pmid.strip()
            url_template = f"https://www.ncbi.nlm.nih.gov/research/pubtator3-api/publications/export/biocxml?pmids={pmid}&full=true"

            response = requests.get(url_template)

            if response.status_code == 200:
                # Parse XML content
                root = ET.fromstring(response.content)
                text_accumulator = ''
                for passage in root.iter('passage'):
                    # Check if the text contains more than one word and is not None
                    text_element = passage.find('text')
                    if text_element is not None:
                        text = text_element.text.strip()
                        text_accumulator += text + ' '  # Concatenate text
                if contains_multiple_words(text_accumulator):
                    pmid_texts[pmid] = text_accumulator
                else:
                    failed_file.write(pmid + '\n')
            else:
                print(f"Failed to retrieve data for PMID: {pmid}")
                failed_file.write(pmid + '\n')

            # Periodically dump data to JSON file
            if len(pmid_texts) % 100 == 0:
                with open(output_file_path, 'w', encoding='utf-8') as outfile:
                    json.dump(pmid_texts, outfile, indent=4)

            # Update progress bar
            progress_bar.update(1)

            # Limit requests to 3 per second
            if progress_bar.n % 3 == 0:
                elapsed_time = time.time() - start_time
                if elapsed_time < 1:
                    time.sleep(1 - elapsed_time)
                start_time = time.time()

        progress_bar.close()

# Dump the remaining data to the JSON file
with open(output_file_path, 'w', encoding='utf-8') as outfile:
    json.dump(pmid_texts, outfile, indent=4)
```
